```python
import csv
from pprint import pprint
```

```python
mpg_file = r'mpg.csv'
```

```python
with open(mpg_file) as csv_file:
    mpg = list(csv.DictReader(csv_file))
```

```python
mpg
```

```python
cylinders = set(cyl['cyl'] for cyl in mpg)
```

```python
cylinders # The unique values for the number of cylinders the cars have in the dataset.
```

<br>
mpg: miles per gallon;
class: car classification;
cty: city mpg;
cyl: # of cylinders;
displ: engine displacement in liters;
drv: f = front-wheel drive, r = rear wheel drive, 4 = 4wd;
fl: fuel (e = ethanol E85, d = diesel, r = regular, p = premium, c = CNG);
hwy: highway mpg;
manufacturer: automobile manufacturer;
model: model of car;
year: model year4;

<br>
Grouping cars by number of cylinder and finding average city miles per gallon

```python
cty_mpg_by_cyl = []
%precision 2

for c in cylinders:
    sum_cty = 0
    count_cyl = 0
    for item in mpg:
        if item['cyl'] == c:
            count_cyl += 1
            sum_cty += float(item['cty'])
    cty_mpg_by_cyl.append((c, sum_cty / count_cyl))
cty_mpg_by_cyl.sort(key=lambda x: x[0])
```

```python
cty_mpg_by_cyl
```

```python
vehicle_classes = set(cls_vhcl['class'] for cls_vhcl in mpg ) # Unique vehicle classes in the dataset.
```

```python
vehicle_classes
```

<br>
Average highway miles per gallon.

```python
hwy_mpg_by_class = []

for vhcl_class in vehicle_classes:
    sum_hwy = 0
    count_v = 0
    for item in mpg:
        if item['class'] == vhcl_class:
            count_v += 1
            sum_hwy += float(item['hwy'])
    hwy_mpg_by_class.append((vhcl_class, sum_hwy / count_v))
hwy_mpg_by_class.sort(key=lambda x: x[1])
```
